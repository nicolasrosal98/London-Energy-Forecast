## 2. Preparing Data for Modeling

```python
#function to create test and train set
def train_test(dataframe, border):
    '''
    Splits dataframe into train and test sets given an index splitting point. 
    
    Arguments:
    dataframe -- dataframe to split. 
    border -- splitting index. 
    
    Returns:
    train set, test set
    '''
    
    return dataframe.loc[:border], dataframe.loc[border:]
```

```python
def scaler(train_df, test_df):
    '''
    Creates MinMaxScaler object and scales train set and test sets.
    
    Arguments:
    train_df -- train set dataframe
    test_df -- test set dataframe
    
    Returns:
    scaler -- scaler object
    scaled_train -- scaled train set dataframe
    scaled_test -- scaled test set dataframe
    
    '''
    
    scaler = MinMaxScaler(feature_range=(0,1))
    scaled_train = scaler.fit_transform(train_df).flatten()
    scaled_test = scaler.transform(test_df).flatten()
    
    return scaler, scaled_train, scaled_test
```

```python
#function to create timeseries X and y (Adapted from: Jason Brownlee)
def create_timeseries(sequence, timestepsx, timestepsy = 0, n_features = 1):
    
    '''
    Creates X and Y (target) from a time-series dataframe sequence for supervised learning. 
    
    Arguments:
    sequence -- Time-series dataframe sequence
    timestepsx -- Number of timesteps assigned to X (number of inputs for prediction).
    timestepsy -- Number of timesteps assigned to Y (number of targets to predict). If timestepsy = 0, 
                  only X will be created. Default is 0. 
    n_features -- Number of features of the dataframe used. Default is 1. 
    
    Returns:
    X -- 3D array (samples, timesteps, n_features) containing X values for model. 
    Y -- 3D array (samples, timesteps, n_features) containing Y values for model. 
    
    '''
    
    X, Y = [], []
    
    if timestepsy != 0: 
        for i in range(len(sequence)):
            #Define end of input and output sequences
            inp_end_idx = i + timestepsx
            out_end_idx = inp_end_idx + timestepsy
        
            #Break if end of output sequence is out of the sequence length
            if out_end_idx > len(sequence):
                break
        
            #Create seq_x and seq_y
            seq_x = sequence[i:(inp_end_idx)]
            X.append(seq_x)
            seq_y = sequence[(inp_end_idx): (out_end_idx)]
            Y.append(seq_y)
                        
    else:
        for i in range(len(sequence) - timestepsx + 1):
            inp_end_idx = i + timestepsx
            seq_x = sequence[i:(inp_end_idx)]
            X.append(seq_x)
            
    
    X = np.array(X)
    X = X.reshape((X.shape[0], timestepsx, n_features))
    Y = np.array(Y)
        
    return X, Y
        
#####ADAPTED FROM: https://machinelearningmastery.com/how-to-develop-lstm-models-for-time-series-forecasting/
```

## 3. Modelling

```python
#Forecast n days into the future

def forecast(model_sel, forecast_data, timestepsx, n_forecast, days_forward, n_features = 1):

    '''
    Function that returns a desired number of forecasted values in a list. 
    
    Arguments:
    model_sel -- LSTM model used to make forecasts. 
    forecast_data -- time-series dataframe used to generate forecast (X). 
    timestepsx -- Number of timesteps assigned to X (number of inputs for prediction). 
                  Used in create_timeseries function. Since we want to create only the X, we use the default value
                  for timestepsy. 
    n_forecast -- Number of Ys to be forecasted.
    days_forward -- Number of values in each Y selected as predictions of the current batch. These will also be 
                    appended as new values for the next batch used to forecast (recurrent forecasting). 
    n_features --  Number of features of the dataframe used. Default is 1. 
    
    Returns:
    pred_list -- list containing forecasted values (length = n_forecast * days_forward). 
    
    '''

    pred_list = []
    train_forecast, _ = create_timeseries(forecast_data, timestepsx, n_features = n_features)
    batch = train_forecast[-1:]
    
    for i in range(n_forecast):
        pred_list.append(model_sel.predict(batch)[0][:days_forward])
        batch = np.append(batch[:,days_forward:,:], [pred_list[i]] ,axis=1)
        
    return pred_list
```

```python
def forecast_visualization(train_df, test_df, scaler, forecast, n_forecast, days_forward):

    '''
    Returns a visualization of the forecasted values, the input values and the test values. 
    
    Arguments:
    train_df -- training data set.
    test_df -- testing data set. 
    scaler -- scaler used to scale the datasets.
    forecast -- prediction list generated with the forecast function. 
    n_forecast -- Number of Ys to be forecasted 
    days_forward -- Number of values in each Y selected as predictions.
    
    Return:
    df_proj -- dataframe that contains the predictions generated and the test data. This will be used to compare 
               the values predicted vs. the real values. 
               
    
    '''

    idx = n_forecast * days_forward
    
    #Creating new dates
    add_dates = [(pd.to_datetime(train_df.index)[-1] + DateOffset(days = x)).date().strftime("%Y-%m-%d") for x in range(0, idx)]
    future_dates = pd.DataFrame(index=add_dates[0:],columns=train_df.columns)
    future_dates.index = pd.to_datetime(future_dates.index)
    
    #Creating projections df
    pred_array = np.array(forecast) 
    pred_array = pred_array.reshape(pred_array.shape[0], pred_array.shape[1])
    pred_array = scaler.inverse_transform(pred_array).flatten()
    
    df_predict = pd.DataFrame(pred_array,
                              index=future_dates[-idx:].index, columns=['Prediction'])
    df_predict.index = df_predict.index.date
    test_df.index = pd.to_datetime(test_df.index).date
    df_proj = pd.concat([test_df, df_predict], axis=1)
    
    #Plotting
    plt.figure(figsize=(20, 5))
    plt.plot(pd.to_datetime(df_proj.index[:idx]), df_proj['energy_sum'][:idx])
    plt.plot(pd.to_datetime(df_proj.index), df_proj['Prediction'], color='r')
    plt.plot(pd.to_datetime(train_df.index[-idx:]), train_df['energy_sum'][-idx:], color='g')
    
    plt.show()

    return df_proj
```

It was found that a TimeDistributed Layer reduced significantly the MAE of the predictions when performing forecasting tasks. For this reason this model was created with two different output layers that can change to fit the assigment to be performed (Trend recognition or Forecasting).

```python
def model(dataset, time_distributed = True, n_features_output = 1):
    
    '''
    Returns an LSTM model with 5 hidden LSTM layers and 5 dropouts (0.25), optimised with RMSprop. Moreover,
    the model instance can be created with a time-distributed dense layer or a dense layer. 
    
    Arguments:
    dataset -- dataset that will be used to specify the input shape of the model. 
    time_distributed -- parameter that changes the output layer between a time-distributed dense layer (if true)
                        and a dense layer (if false).
    n_features_output -- number of features we are predicting in the output. 
    
    Return:
    model -- returns a model instance. 
    
    '''

    np.random.seed(123)

    #LSTM Architecture

    model = Sequential()

    #Hidden Layer 1
    model.add(LSTM(units = 64, return_sequences = True, input_shape = (dataset.shape[1], dataset.shape[2])))

    #Dropout Layer 1
    model.add(Dropout(rate = 0.25))

    #Hidden Layer 2
    model.add(LSTM(units = 64, return_sequences = True))
    
    #Dropout Layer 2
    model.add(Dropout(rate = 0.25))

    #Hidden Layer 3
    model.add(LSTM(units = 64, return_sequences = True))

    #Dropout Layer 3
    model.add(Dropout(rate = 0.25))

    #Hidden Layer 4
    model.add(LSTM(units = 64, return_sequences = True))

    #Dropout Layer 4
    model.add(Dropout(rate = 0.25))
    
    if time_distributed:
        
        #Hidden Layer 5
        model.add(LSTM(units = 64, return_sequences = True))
        
        #Output Layer 
        model.add(TimeDistributed(Dense(n_features_output)))
        
    else: 
        #Hidden Layer 5
        model.add(LSTM(units = 64))
        
        #Output Layer 
        model.add(Dense(n_features_output))

    #Compile
    model.compile(loss = 'mean_squared_error', optimizer = 'RMSprop', metrics = ['mean_absolute_error'])

    #Summary
    print(model.summary())
    
    return model
```

### Model Using All Features 

The procedure performed in "Total Energy Use in London" will be replicated with all the features of the datasets in order to test if it can improve forecasting performance.

```python
def scalera(train_df, test_df):
    '''
    Creates MinMaxScaler object and scales train set and test sets. Moreover, it creates a scaler object for 
    the target column. 
    
    Arguments:
    train_df -- train set dataframe
    test_df -- test set dataframe
    
    Returns:
    scaler -- scaler object for train and test.
    scaler_target -- scaler object for target column.
    scaled_train -- scaled train set dataframe
    scaled_test -- scaled test set dataframe
    
    
    '''
    
    scaler = MinMaxScaler(feature_range=(0,1))
    scaled_train = scaler.fit_transform(train_df)
    scaled_test = scaler.transform(test_df)
    
    scaler_target = MinMaxScaler(feature_range=(0,1))
    scaled_target = scaler_target.fit(pd.DataFrame(train_df.iloc[:,-1]))
    
    return scaler, scaler_target, scaled_train, scaled_test
```

```python
#Scaling Data
scaler_features, scaler_target, scaled_featurestrain, scaled_featurestest = scalera(energytrain_df[['Avg.Dawn','Avg.Morning_Afternoon',
                                                                                                    'Avg.Evening','Avg.Night', 'energy_sum']],
                                                              energytest_df[['Avg.Dawn','Avg.Morning_Afternoon',
                                                                                'Avg.Evening','Avg.Night', 'energy_sum']])
```
